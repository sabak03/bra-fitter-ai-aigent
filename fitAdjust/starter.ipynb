Basic Measurements UK and in Inches

```python
cups = ["A","B","C","D","E","F","G","H","I","J","K","L","M","N","O","P"]
double = ["D","F","H","J","K"]
sisters = ["AA","A","B","C","D","DD","E","F","FF","G","H","HH","I","J","JJ","K","KK","L","M","N","O","P"]
```

```python
def get_basic_size():
    under = int(input("Underbust measuerment(in inches):"))
    full = int(input("Full bust measuerment(in inches):"))
    if under % 2 == 0:
        band = under
        odd = False
    else:
        band = under + 1
        band = band
        odd = True
    cup = full-under
    cup = cups[cup-1]
    if odd and cup in double:
        cup = cup+cup
    size = str(band)+cup
    return size, band, cup
```

```python
size, band, cup = get_basic_size()
print(size)
```

Sister Sizing

```python
def get_sister_sizes():
    cupIndex = sisters.index(cup)
    sister = [(str(band+4)+sisters[cupIndex-2]),(str(band+2)+sisters[cupIndex-1]),size,(str(band-2)+sisters[cupIndex+1]),(str(band-4)+sisters[cupIndex+2])]
    return sister
```

```python
sister = get_sister_sizes()
print(sister)
```
